### 2. (4 points) 
Explain why the optimal market/limit order decision depends in general on quantity available on both sides of the order book in the Parlour model.

One assumption in Parlour model is that if the limit order has 0 probability to be executed, limit order will not be submited. So if the quantity available on both sides is very large, the probability of executing a limit order (placed at the end of waiting queue) is very low and the expected waiting time for execution is long requiring $\beta$ to be very high. So if queue size is large, it will be more likely to submit a market order than a limit order

## The Bouchaud-M&eacute;zard-Potters order book approximation

### 4. (16 points)

When $\lambda(u)=\alpha$,  a constant, the BMP order density becomes

\begin{eqnarray*}
\rho_{BMP}(\Delta) &=& \frac{\alpha}{\delta}\,\left\{e^{-\Delta/{\tilde \Delta}}\,\int_0^{\Delta/{\tilde\Delta}}\,\sinh(u)\,du + \sinh(\Delta/{\tilde \Delta})\,\int_{\Delta/{\tilde\Delta}}^\infty\,e^{-u}\,du\right\} \\
&=&\frac{\alpha}{\delta}\,\left\{1 - e^{-\Delta/{\tilde \Delta}}  \right\}
\end{eqnarray*}

where the characteristic distance

$$
\tilde \Delta=\sqrt{\frac{\sigma^2}{2\,\delta}}.
$$

```r
#a)
#updated code
logging<- F #set logging to F to speed up running
pos=21 #store the mid position index
alpha<- 1 #set alpha
mu<- 10 #set mu
delta<-0.2
numEvents<- 100000 #set numEvents
initializeBook5() #initialize order book
band<- 20 #set band
aveBookShape<- bookShape(band)/numEvents #initialize average book shape
countLimitBuy<- rep(0,30) #store count of limit buys
countLimitSell<- rep(0,30) #store count of limit sells
nb <<- sum(book$buySize[book$Price>=(bestOffer()-L)]); # Number of cancelable buy orders
ns <<- sum(book$sellSize[book$Price<=(bestBid()+L)]); # Number of cancelable sell orders
eventRate <- nb*delta+ns*delta + mu +2*L*alpha;
probEvent <- c(L*alpha,L*alpha,nb*delta,ns*delta,mu/2,mu/2)/eventRate;
for (i in 1:numEvents){
  m <- sample(1:6, 1, replace = TRUE, probEvent); #Choose event type
  price<-0
  level=0
  if (m==1){ #selected event is limitbuy
    level=pick(L) #set level
    price <- (bestOffer()-level) #set price
    countLimitBuy[level]<-countLimitBuy[level]+1 #increment limit buy count at this level
  }
  if(m==2){
    level=pick(L) #selected event is limit sell and set level
    price<-(bestBid()+level) #set price
    countLimitSell[level]<-countLimitSell[level]+1 #increment limit sell count at this level
  }
  switch(m,
         limitBuyOrder(price),
         limitSellOrder(price),
         cancelBuyOrder(),
         cancelSellOrder(),
         marketBuyOrder(),
         marketSellOrder()
  );
  aveBookShape<- aveBookShape+bookShape(band)/numEvents
}
#b) plot order shape
plot(-band:band,aveBookShape,main=c("average book shape"), xlab="Relative Price",ylab="Shares",col="red",type="b")
#c)
total_prob=rep(0,60)
for (i in 1:L){
    total_prob[i]=countLimitSell[i]/numEvents
    total_prob[i+30]=countLimitBuy[31-i]/numEvents
}
print(countLimitSell/numEvents)
print(countLimitBuy/numEvents)
plot(1:60,total_prob,main=c("probability of limit order at various levels"),xlab="level",ylab="probability",col="red",type="b")
```

(a) Derive an expression in terms of the Student-$t$ cumulative density function for the ratio of arrival rates of limit to market orders as a function of the spread $s$.  How does this ratio vary with $s$?

$$P(LB)=0.5\times F(\frac{s}{2})$$
$$P(LS)=P(LB)$$

$$P(MS)=P(MB)$$
Thus, $$Limit/Market=F(\frac{s}{2})/(1-F(\frac{s}{2}))$$
According to the plot below, the larger the spread, it is more likely to come a limit order than a market order.

(b) Relative to the zero-intelligence model, does this empirically observed strategic behavior make the spread revert faster or more slowly?

Answer: In ZI model, arrival of limit and market orders does not vary with spread. But with this observed behavior, when spread is larger, limit order arrives more frequently than market order. Limit order can narrow the spread and market order can widen the spread. So such behavior makes the spread revert faster.
